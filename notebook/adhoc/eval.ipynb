```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
!ls ../../log/
```

```python
df = pd.read_csv('../../log/lstm&batchsize=128&lr=0.00003&mo=0.9&epochs=100.csv')
display(df.head())
fig = plt.figure(figsize=(18, 6))

# accuracy
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(df['epoch'], df['train_acc'])
ax1.plot(df['epoch'], df['test_acc'])
ax1.set_title('acc')

# loss
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(df['epoch'], df['train_loss'])
ax2.plot(df['epoch'], df['test_loss'])
ax2.set_title('loss')
plt.show()
```

- バッチサイズ大きいと局所解に陥りやすい
- バッチサイズ小さいと学習遅い
- xx だと過学習しやすい
- lr 大きいと学習曲線ガタガタになる？
- lr 小さいと 滑らかすぎて収束しない

```python
df = pd.read_csv('../../log/emb=random&edim=300&model=lstm&hdim=100&batchsize=032&optim=sgd&lr=0.10000&mo=0.9&epochs= 100&shuffle=True.csv')
display(df.head())
fig = plt.figure(figsize=(18, 6))

# accuracy
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(df['epoch'], df['train_acc'])
ax1.plot(df['epoch'], df['test_acc'])
ax1.set_title('acc')

# loss
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(df['epoch'], df['train_loss'])
ax2.plot(df['epoch'], df['test_loss'])
ax2.set_title('loss')
plt.show()
```
